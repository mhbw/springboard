# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
```

****
## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

## A. Do you agree with the above analysis and recommendations? Why or why not?

On it's face I have some problems here, in that I think this likely falls pray to the [law of small numbers](http://stats.org.uk/statistical-inference/TverskyKahneman1971.pdf) which dictates that "we have a strong bias towards believing that small samples closely resemble the population from which they are drawn". 

Essentially a small blip in the population at a small hospital will create a large deviation. Backing this up, if you look at the top and bottom of 10% they are overwhelmingly the smaller hospitals.

All that is to say that I don't think it can be totally chalked up to size. There could be something else going on such as the different sets of services offered. Requiring that they upgraded resource allocation might not be the correct answer, as they could be simply offering specialist services that are risky or to a higher-risk population.

## 1. Setup an appropriate hypothesis test.

I propose that the actual risk comes more from the types of surgery rather than the size of the hospital.   Therefore, we'll say the null-hypothesis is that there is no difference based on proceedure. 



First, some general exploration and analysis:

```python
##clean_hospital_read_df.loc['Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)

print("old rows: ",len(clean_hospital_read_df))


clean_hospital_read_df.sort_values(by='Number of Discharges', ascending=False, inplace=True)
## dropping na values in terms of readmission
clean_hospital_read_df = clean_hospital_read_df.dropna(subset=['Number of Discharges', 'Excess Readmission Ratio'])

print("new rows: ",len(clean_hospital_read_df))
print(clean_hospital_read_df.head())

```

Quick glance the surgeries themselves:

```python
clean_hospital_read_df.boxplot(column='Excess Readmission Ratio',by='Measure Name')

```

```python
import scipy.stats as stats
import researchpy as rp
import statsmodels.api as sm
from statsmodels.formula.api import ols

rp.summary_cont(clean_hospital_read_df['Excess Readmission Ratio'].groupby(clean_hospital_read_df['Measure Name']))
```

This would show the indication that the measures ahve about the same rate of readmission, right on about 1.0, but the variance here across surgeries is noteable, particularly with hip and kneee replacememts. as you can see the standard deviation is much higher there, as is the over all mean, which is significantly higher.  

## 2.  Compute and report the observed significance value (or p-value).
## 3.  Report statistical significance for  𝛼 = .01.

Let's do this via an F-statistic, since we have multiple groups. 

```python
AMI = clean_hospital_read_df[clean_hospital_read_df['Measure Name']=='READM-30-AMI-HRRP']
COPD = clean_hospital_read_df[clean_hospital_read_df['Measure Name']=='READM-30-COPD-HRRP']
HF = clean_hospital_read_df[clean_hospital_read_df['Measure Name']=='READM-30-HF-HRRP']
PN = clean_hospital_read_df[clean_hospital_read_df['Measure Name']=='READM-30-PN-HRRP']
## confusingly CABG doesn't show up in the data set. unclear why?
HIP_KNEE = clean_hospital_read_df[clean_hospital_read_df['Measure Name']=='READM-30-HIP-KNEE-HRRP']

print('number AMI :', len(AMI))
print('number COPD :', len(COPD))
print('number HF :', len(HF))
print('number PN :', len(PN))
print('number Hip Knee :', len(HIP_KNEE))
```

```python
stats.f_oneway(AMI['Excess Readmission Ratio'], COPD['Excess Readmission Ratio'], HF['Excess Readmission Ratio'], PN['Excess Readmission Ratio'], HIP_KNEE['Excess Readmission Ratio'])
```

Our F-statistic is 19.56, with a p-value of < .001. This is well below 𝛼 = .01 so we can reject the null hypothesis; clearly the procedure type has a sigificant impact on readmission rates.

#### Construct another plot that conveys the same information in a more direct manner.

First I choose to try an ECDF chart which we have used frequently thus far. 

```python
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y

x_temp, y_temp = ecdf(clean_hospital_read_df['Excess Readmission Ratio'])

plt.plot(x_temp, y_temp,marker='.', linestyle='none')
plt.xlabel('Excess Readmission Ratio')
plt.ylabel('ECDF')
plt.show()
```

This is far simpler, but is actually perhaps too simple. After all, you can see there is a steep curve, but we have little information as to what that means. Therefore I thought we might use a charting of the log of the values. I think that's important, becuase it smooths the data and reduces the effect of the large numbers. I choose to use log 10 for the number of discharges given that they are larger numbers, and 2 of the excess rate since the values cluster around 1. 

```python
x_log = np.log10(x)
y_log = np.log2(y)

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x_log, y_log,alpha=0.2)

ax.set_xlim([min(x_log), max(x_log)])
ax.set_xlabel('log of discharges', fontsize=12)
ax.set_ylabel('Excess log rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of log rate of discharges vs. excess log rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```
